```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
from itertools import combinations
import os

# =============================================================================
# CONFIGURAÇÃO DO ARQUIVO CSV
# =============================================================================
print("=" * 70)
print("ANÁLISE DE EFICIÊNCIA ENERGÉTICA DE MODELOS LLM")
print("=" * 70)

# Nome do arquivo CSV a ser lido (ALTERE PARA O SEU CAMINHO)
ARQUIVO_CSV = "/content/Experimento_Eficiencia_Energetica.csv"

# =============================================================================
# 1. CARREGAR O ARQUIVO CSV
# =============================================================================
print(f"Carregando arquivo: {ARQUIVO_CSV}")

# Leitura e limpeza dos dados
df = pd.read_csv(ARQUIVO_CSV, sep=';')

print(f"✓ Dados carregados: {len(df)} linhas, {len(df.columns)} colunas")
print(f"✓ Colunas disponíveis: {list(df.columns)}")
print("\nPrimeiras linhas do DataFrame:")
print(df.head())

# =============================================================================
# 2. CONVERTER COLUNAS PARA NUMÉRICO
# =============================================================================
print("\n" + "-" * 70)
print("CONVERTENDO COLUNAS PARA VALORES NUMÉRICOS")
print("-" * 70)

# Converter colunas para numeric, handling potential comma decimal separators

df['Consumo_Energia_Token'] = df['Consumo_Energia_Token'].str.replace(',', '.').astype(float)
df['Tempo_Execucao_Token'] = df['Tempo_Execucao_Token'].str.replace(',', '.').astype(float)

print(f"✓ Conversão concluída")
print(f"  Consumo_Energia_Token: {df['Consumo_Energia_Token'].dtype}")
print(f"  Tempo_Execucao_Token: {df['Tempo_Execucao_Token'].dtype}")

# =============================================================================
# 3. CALCULAR MÉTRICAS POR TOKEN
# =============================================================================
print("\n" + "-" * 70)
print("CALCULANDO MÉTRICAS POR TOKEN")
print("-" * 70)

# Calcular métricas por token
TOKENS = 256
df["energia_token_j"] = df["Consumo_Energia_Token"] / TOKENS
df["tempo_token_s"]   = df["Tempo_Execucao_Token"] / TOKENS

print(f"✓ Métricas calculadas (TOKENS = {TOKENS})")
print(f"\nEstatísticas dos dados:")
print(f"  Número de observações: {len(df)}")
print(f"  Modelos: {df['Modelo'].unique().tolist()}")
print(f"\nEnergia por token (J):")
print(f"  Média: {df['energia_token_j'].mean():.6f}")
print(f"  Mínimo: {df['energia_token_j'].min():.6f}")
print(f"  Máximo: {df['energia_token_j'].max():.6f}")
print(f"  Mediana: {df['energia_token_j'].median():.6f}")
print(f"\nTempo por token (s):")
print(f"  Média: {df['tempo_token_s'].mean():.6f}")
print(f"  Mínimo: {df['tempo_token_s'].min():.6f}")
print(f"  Máximo: {df['tempo_token_s'].max():.6f}")

# Definir coluna do modelo
modelo_col = 'Modelo'  # A coluna se chama 'Modelo' no seu CSV

# =============================================================================
# 4. ANÁLISE DE CORRELAÇÃO
# =============================================================================
print("\n" + "=" * 70)
print("ANÁLISE DE CORRELAÇÃO (PEARSON)")
print("=" * 70)

# Calculate Pearson correlation
corr_pearson, p_pearson = stats.pearsonr(
    df["energia_token_j"],
    df["tempo_token_s"]
)

print(f"Pearson correlation coefficient (r): {corr_pearson:.4f}")
print(f"P-value: {p_pearson:.4f}")

# Spearman correlation (opcional)
corr_spearman, p_spearman = stats.spearmanr(df["energia_token_j"], df["tempo_token_s"])
print(f"\nSpearman correlation (rho): {corr_spearman:.4f}")
print(f"P-value: {p_spearman:.4f}")

# =============================================================================
# 5. TESTES DE NORMALIDADE
# =============================================================================
print("\n" + "=" * 70)
print("TESTES DE NORMALIDADE (SHAPIRO-WILK)")
print("=" * 70)

print("Teste de normalidade para energia por token por modelo:")
for modelo in df[modelo_col].unique():
    energia = df[df[modelo_col] == modelo]["energia_token_j"]
    stat, p = stats.shapiro(energia)
    normal = "✓ Normal" if p > 0.05 else "✗ Não normal"
    print(f"  {modelo:20}: p-value = {p:.4f} ({normal})")

# =============================================================================
# 6. INTERVALOS DE CONFIANÇA 95%
# =============================================================================
print("\n" + "=" * 70)
print("INTERVALOS DE CONFIANÇA 95% POR MODELO")
print("=" * 70)

def ic_95(series):
    """Calcula média e intervalo de confiança 95% usando t-distribuição"""
    mean = series.mean()
    sem = stats.sem(series)
    ic = stats.t.interval(0.95, len(series)-1, loc=mean, scale=sem)
    return mean, ic

ic_results = {}
for modelo in df[modelo_col].unique():
    energia = df[df[modelo_col] == modelo]["energia_token_j"]
    mean, ic = ic_95(energia)
    ic_results[modelo] = {'mean': mean, 'ci_95': ic}
    print(f"{modelo:20}: Média = {mean:.6f} J/token")
    print(f"                    IC 95% = [{ic[0]:.6f}, {ic[1]:.6f}] J/token")

# =============================================================================
# 7. TESTES T PAREADOS
# =============================================================================
print("\n" + "=" * 70)
print("TESTES T PAREADOS (Welch's t-test)")
print("=" * 70)

modelos_list = df[modelo_col].unique()
for i in range(len(modelos_list)):
    for j in range(i+1, len(modelos_list)):
        m1 = df[df[modelo_col] == modelos_list[i]]["energia_token_j"]
        m2 = df[df[modelo_col] == modelos_list[j]]["energia_token_j"]
        stat, p = stats.ttest_ind(m1, m2, equal_var=False)
        sig = "✓ DIFERENÇA SIGNIFICATIVA" if p < 0.05 else "✗ sem diferença significativa"
        print(f"{modelos_list[i]:15} vs {modelos_list[j]:15}: p = {p:.4f} ({sig})")

# =============================================================================
# 8. GRÁFICO 1: BOXPLOT - ENERGIA POR TOKEN POR MODELO
# =============================================================================
print("\n" + "=" * 70)
print("GRÁFICO 1: Boxplot - Energia por token por modelo")
print("=" * 70)

plt.figure(figsize=(12, 8))
df.boxplot(column="energia_token_j", by=modelo_col, rot=45, grid=False)

# Configurar escala do eixo Y baseada nos dados
y_min = df["energia_token_j"].min() * 0.98
y_max = df["energia_token_j"].max() * 1.02
plt.ylim(y_min, y_max)

# Configurar ticks
n_ticks = 8
y_ticks = np.linspace(y_min, y_max, n_ticks)
plt.yticks(y_ticks, [f"{y:.4f}" for y in y_ticks])

plt.title("Energia por token por modelo", fontsize=14, fontweight='bold')
plt.suptitle("")
plt.ylabel("Energia por token (J)", fontsize=12)
plt.xlabel("Modelo", fontsize=12)

# Adicionar linha da média geral
media_geral = df["energia_token_j"].mean()
plt.axhline(y=media_geral, color='red', linestyle='--', alpha=0.7,
           linewidth=2, label=f'Média geral: {media_geral:.4f} J')
plt.legend(loc='upper right')

plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig("fig1_boxplot_energia.png", dpi=300, bbox_inches='tight')
plt.show()
print("✓ Gráfico 1 salvo como 'fig1_boxplot_energia.png'")

# =============================================================================
# 9. GRÁFICO 2: SCATTER PLOT - CORRELAÇÃO
# =============================================================================
print("\n" + "=" * 70)
print("GRÁFICO 2: Scatter plot - Correlação tempo vs energia")
print("=" * 70)

plt.figure(figsize=(12, 8))

# Criar scatter plot
plt.scatter(df["tempo_token_s"], df["energia_token_j"],
            alpha=0.6, edgecolors='black', linewidth=0.5, s=50)

plt.xlabel("Tempo por token (s)", fontsize=12)
plt.ylabel("Energia por token (J)", fontsize=12)
plt.title("Correlação entre tempo e energia por token", fontsize=14, fontweight='bold')

# Adicionar linha de regressão
if len(df) > 1:
    x = df["tempo_token_s"]
    y = df["energia_token_j"]
    mask = ~np.isnan(x) & ~np.isnan(y)
    x_clean = x[mask]
    y_clean = y[mask]

    if len(x_clean) > 1:
        slope, intercept, r_value, p_value, std_err = stats.linregress(x_clean, y_clean)

        x_line = np.array([x_clean.min(), x_clean.max()])
        y_line = slope * x_line + intercept

        plt.plot(x_line, y_line, 'r-', linewidth=2,
                label=f'Regressão: r = {r_value:.3f}, p = {p_value:.4f}')
        plt.legend(loc='best')

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("fig2_scatter_tempo_energia.png", dpi=300, bbox_inches='tight')
plt.show()
print("✓ Gráfico 2 salvo como 'fig2_scatter_tempo_energia.png'")

# =============================================================================
# 10. TEOREMA CENTRAL DO LIMITE (TCL) - ANÁLISE (PRIMEIRO CÓDIGO)
# =============================================================================
print("\n" + "=" * 70)
print("TEOREMA CENTRAL DO LIMITE (TCL) - ANÁLISE")
print("=" * 70)
print("\nO TCL estabelece que, para amostras suficientemente grandes (n ≥ 30),\n"
      "a distribuição das médias amostrais tende à normalidade, independentemente\n"
      "da distribuição original dos dados.\n")

# Configurações TCL
n_simulations = 1000
sample_size = 100
confidence_level = 0.95

# Função para aplicar o TCL via bootstrap
def apply_tcl_bootstrap(data, n_simulations=1000, sample_size=100):
    """
    Aplica o Teorema Central do Limite usando bootstrap.
    """
    means = np.zeros(n_simulations)

    for i in range(n_simulations):
        # Amostra aleatória com reposição
        bootstrap_sample = np.random.choice(data, size=sample_size, replace=True)
        means[i] = np.mean(bootstrap_sample)

    return means

# Analisar cada modelo usando TCL
print("\n" + "=" * 70)
print("APLICAÇÃO DO TCL POR MODELO")
print("=" * 70)

tcl_results = {}
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for idx, modelo in enumerate(df[modelo_col].unique()):
    # Dados do modelo
    energia = df[df[modelo_col] == modelo]["energia_token_j"].values

    print(f"\n{'='*40}")
    print(f"MODELO: {modelo}")
    print(f"{'='*40}")
    print(f"Tamanho da amostra original: {len(energia)} observações")
    print(f"Média original: {np.mean(energia):.6f} J/token")
    print(f"Desvio padrão original: {np.std(energia, ddof=1):.6f} J/token")

    # Aplicar TCL via bootstrap
    bootstrap_means = apply_tcl_bootstrap(energia, n_simulations, sample_size)

    # Estatísticas das médias bootstrap
    mean_of_means = np.mean(bootstrap_means)
    std_of_means = np.std(bootstrap_means, ddof=1)
    sem = std_of_means  # Erro padrão da média

    # Intervalo de confiança (95%)
    z_score = 1.96  # Para 95% de confiança
    ci_lower = mean_of_means - z_score * sem
    ci_upper = mean_of_means + z_score * sem

    tcl_results[modelo] = {
        'mean_original': np.mean(energia),
        'mean_of_means': mean_of_means,
        'std_of_means': std_of_means,
        'ci_95': (ci_lower, ci_upper),
        'relative_error': sem / mean_of_means * 100  # Erro relativo em %
    }

    print(f"\nTCL - Distribuição das Médias Amostrais:")
    print(f"  Média das médias bootstrap: {mean_of_means:.6f} J/token")
    print(f"  Erro padrão da média: {sem:.6f} J/token")
    print(f"  Intervalo de 95% confiança: [{ci_lower:.6f}, {ci_upper:.6f}] J/token")
    print(f"  Erro relativo: {sem/mean_of_means*100:.2f}%")

    # Plotar histograma das médias bootstrap
    ax = axes[idx]
    ax.hist(bootstrap_means, bins=30, edgecolor='black', alpha=0.7,
           density=True, color='skyblue')

    # Adicionar curva normal teórica
    x = np.linspace(min(bootstrap_means), max(bootstrap_means), 100)
    normal_pdf = stats.norm.pdf(x, mean_of_means, std_of_means)
    ax.plot(x, normal_pdf, 'r-', linewidth=2, label='Distribuição Normal')

    # Marcar intervalo de confiança
    ax.axvline(ci_lower, color='green', linestyle='--', alpha=0.7, label='IC 95%')
    ax.axvline(ci_upper, color='green', linestyle='--', alpha=0.7)
    ax.axvline(mean_of_means, color='red', linewidth=2, label='Média')

    ax.set_title(f'{modelo}\nn={len(energia)}', fontsize=11)
    ax.set_xlabel('Energia por token (J)')
    ax.set_ylabel('Densidade')
    ax.legend(fontsize=9)
    ax.grid(True, alpha=0.3)

plt.suptitle('Teorema Central do Limite - Distribuição das Médias Amostrais\n'
            f'(n_simulações={n_simulations}, tamanho_amostra={sample_size})',
            fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('tcl_bootstrap_distribuicoes.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Gráfico TCL salvo como 'tcl_bootstrap_distribuicoes.png'")

# =============================================================================
# 11. COMPARAÇÃO ENTRE MODELOS USANDO TCL
# =============================================================================
print("\n" + "=" * 70)
print("COMPARAÇÃO ENTRE MODELOS - BASEADA NO TCL")
print("=" * 70)

# Criar DataFrame para comparação
comparison_df = pd.DataFrame({
    'Modelo': tcl_results.keys(),
    'Média_Original': [tcl_results[m]['mean_original'] for m in tcl_results],
    'Média_TCL': [tcl_results[m]['mean_of_means'] for m in tcl_results],
    'Erro_Padrão': [tcl_results[m]['std_of_means'] for m in tcl_results],
    'CI_Inferior': [tcl_results[m]['ci_95'][0] for m in tcl_results],
    'CI_Superior': [tcl_results[m]['ci_95'][1] for m in tcl_results]
}).sort_values('Média_TCL')

print("\nRanking de eficiência energética (menor = mais eficiente):")
print(comparison_df[['Modelo', 'Média_TCL', 'Erro_Padrão']].to_string(index=False))

# Plotar comparação com barras de erro
plt.figure(figsize=(12, 6))
x_pos = np.arange(len(comparison_df))
plt.errorbar(x_pos, comparison_df['Média_TCL'],
             yerr=comparison_df['Erro_Padrão'],
             fmt='o', capsize=5, capthick=2,
             color='darkblue', markersize=8, linewidth=2)

# Adicionar intervalo de confiança como área sombreada
for i, row in comparison_df.iterrows():
    plt.fill_betweenx(y=[row['CI_Inferior'], row['CI_Superior']],
                      x1=i-0.2, x2=i+0.2, alpha=0.2, color='blue')

plt.xticks(x_pos, comparison_df['Modelo'], rotation=45, ha='right')
plt.ylabel('Energia por token (J)')
plt.title('Comparação de Eficiência Energética entre Modelos\n'
          '(Barras de erro = ±1 EP, área sombreada = IC 95% baseado no TCL)',
          fontsize=12, fontweight='bold')
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('tcl_comparacao_modelos.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Gráfico de comparação TCL salvo como 'tcl_comparacao_modelos.png'")

# =============================================================================
# 12. TESTES DE HIPÓTESE USANDO TCL
# =============================================================================
print("\n" + "=" * 70)
print("TESTES DE HIPÓTESE - BASEADOS NO TCL")
print("=" * 70)

def compare_models_tcl(model1, model2, tcl_results, alpha=0.05):
    """
    Compara dois modelos usando TCL para teste de hipóteses
    """
    m1 = tcl_results[model1]
    m2 = tcl_results[model2]

    # Diferença das médias
    diff = m1['mean_of_means'] - m2['mean_of_means']

    # Erro padrão da diferença
    se_diff = np.sqrt(m1['std_of_means']**2 + m2['std_of_means']**2)

    # Estatística Z (assumindo normalidade pelo TCL)
    z_stat = diff / se_diff

    # Valor p bilateral
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

    # Decisão
    significant = p_value < alpha

    return {
        'model1': model1,
        'model2': model2,
        'diff': diff,
        'z_stat': z_stat,
        'p_value': p_value,
        'significant': significant
    }

# Comparar todos os pares
print("\nComparações pareadas (teste Z baseado no TCL):")
comparisons = []
for model1, model2 in combinations(df[modelo_col].unique(), 2):
    result = compare_models_tcl(model1, model2, tcl_results)
    comparisons.append(result)

    symbol = "\u2605" if result['significant'] else " "
    print(f"{symbol} {model1:15} vs {model2:15}: "
          f"diff = {result['diff']:+.6f}, "
          f"z = {result['z_stat']:+.2f}, "
          f"p = {result['p_value']:.4f}")

# =============================================================================
# 13. CONCLUSÕES BASEADAS NO TCL
# =============================================================================
print("\n" + "=" * 70)
print("CONCLUSÕES BASEADAS NO TEOREMA CENTRAL DO LIMITE")
print("=" * 70)

print("\n1. Tamanho da amostra:")
print(f"   \u2022 Cada modelo tem aproximadamente {len(df[df[modelo_col] == df[modelo_col].unique()[0]])} observações")
print(f"   \u2022 Amostra > 30: \u2713 Condição do TCL satisfeita")

print("\n2. Precisão das estimativas (Erro Padrão):")
for modelo in tcl_results:
    ep = tcl_results[modelo]['std_of_means']
    ep_rel = tcl_results[modelo]['relative_error']
    print(f"   \u2022 {modelo:15}: EP = {ep:.6f} ({ep_rel:.1f}%)")

print("\n3. Modelos mais eficientes (menor consumo energético por token):")
for i, row in comparison_df.iterrows():
    rank = i + 1
    print(f"   {rank}\u00ba - {row['Modelo']:15}: {row['Média_TCL']:.6f} \u00b1 {row['Erro_Padrão']:.6f} J/token")

print("\n4. Recomendação baseada em evidência estatística:")
best_model = comparison_df.iloc[0]['Modelo']
best_ci = f"[{comparison_df.iloc[0]['CI_Inferior']:.6f}, {comparison_df.iloc[0]['CI_Superior']:.6f}]"
print(f"   \u2022 {best_model} é o mais eficiente com 95% de confiança")
print(f"   \u2022 Consumo médio: {comparison_df.iloc[0]['Média_TCL']:.6f} J/token")
print(f"   \u2022 Intervalo de confiança: {best_ci} J/token")

# =============================================================================
# 14. GRÁFICO 3: SCATTER PLOT COLORIDO POR MODELO
# =============================================================================
print("\n" + "=" * 70)
print("GRÁFICO 3: Scatter plot colorido por modelo")
print("=" * 70)

plt.figure(figsize=(14, 8))

# Cores e marcadores
colors = plt.cm.tab10(np.arange(len(df[modelo_col].unique())))
markers = ['o', 's', '^', 'D', 'v', 'p', '*', 'X']

for idx, modelo in enumerate(df[modelo_col].unique()):
    modelo_data = df[df[modelo_col] == modelo]

    plt.scatter(modelo_data["tempo_token_s"],
                modelo_data["energia_token_j"],
                c=[colors[idx]],
                marker=markers[idx % len(markers)],
                label=modelo,
                alpha=0.7,
                edgecolors='black',
                linewidth=0.5,
                s=80)

plt.xlabel("Tempo por token (s)", fontsize=12)
plt.ylabel("Energia por token (J)", fontsize=12)
plt.title("Correlação entre tempo e energia por token (Colorido por modelo)",
          fontsize=14, fontweight='bold')

# Adicionar legenda
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

# Adicionar estatísticas de correlação
plt.text(0.02, 0.98,
         f'Pearson r = {corr_pearson:.3f}\n'
         f'Spearman ρ = {corr_spearman:.3f}\n'
         f'N = {len(df)}',
         transform=plt.gca().transAxes,
         verticalalignment='top',
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("fig3_scatter_colorido_modelos.png", dpi=300, bbox_inches='tight')
plt.show()
print("✓ Gráfico 3 salvo como 'fig3_scatter_colorido_modelos.png'")

# =============================================================================
# 15. RESUMO E CONCLUSÕES FINAIS
# =============================================================================
print("\n" + "=" * 70)
print("RESUMO E CONCLUSÕES FINAIS")
print("=" * 70)

print(f"\n1. Dados analisados:")
print(f"   • Total de observações: {len(df)}")
print(f"   • Número de modelos: {len(df[modelo_col].unique())}")
print(f"   • Observações por modelo: ~{len(df) // len(df[modelo_col].unique())}")

print(f"\n2. Análise de correlação:")
print(f"   • Pearson (r): {corr_pearson:.4f} (p = {p_pearson:.4f})")
print(f"   • Spearman (ρ): {corr_spearman:.4f} (p = {p_spearman:.4f})")

print(f"\n3. Testes de normalidade:")
for modelo in df[modelo_col].unique():
    energia = df[df[modelo_col] == modelo]["energia_token_j"]
    stat, p = stats.shapiro(energia)
    status = "Normal" if p > 0.05 else "Não normal"
    print(f"   • {modelo:20}: p = {p:.4f} ({status})")

print(f"\n4. Ranking de eficiência (menor energia = mais eficiente):")
for i, row in comparison_df.iterrows():
    rank = i + 1
    print(f"   {rank}º - {row['Modelo']:20}: {row['Média_TCL']:.6f} ± {row['Erro_Padrão']:.6f} J/token")

print(f"\n5. Modelo mais eficiente:")
best_model = comparison_df.iloc[0]['Modelo']
best_mean = comparison_df.iloc[0]['Média_TCL']
best_ci_lower = comparison_df.iloc[0]['CI_Inferior']
best_ci_upper = comparison_df.iloc[0]['CI_Superior']
print(f"   • {best_model} (Média: {best_mean:.6f} J/token)")
print(f"   • IC 95%: [{best_ci_lower:.6f}, {best_ci_upper:.6f}] J/token")

print(f"\n6. Testes de hipóteses significativos (p < 0.05):")
sig_comparisons = [c for c in comparisons if c['significant']]
if sig_comparisons:
    for comp in sig_comparisons:
        print(f"   • {comp['model1']} vs {comp['model2']}: "
              f"diff = {comp['diff']:.6f}, p = {comp['p_value']:.4f}")
else:
    print("   • Nenhuma diferença significativa encontrada entre os modelos")

print(f"\n7. Gráficos gerados:")
print("   • fig1_boxplot_energia.png - Boxplot por modelo")
print("   • fig2_scatter_tempo_energia.png - Correlação tempo-energia")
print("   • fig3_scatter_colorido_modelos.png - Correlação colorida por modelo")
print("   • tcl_bootstrap_distribuicoes.png - Distribuições TCL")
print("   • tcl_comparacao_modelos.png - Comparação TCL entre modelos")

# =============================================================================
# 16. SALVAR RESULTADOS
# =============================================================================
print(f"\n" + "=" * 70)
print("SALVANDO RESULTADOS")
print("=" * 70)

# Salvar DataFrame com métricas
df.to_csv("dados_com_metricas.csv", index=False)
print("✓ Dados com métricas salvos em 'dados_com_metricas.csv'")

# Salvar resultados TCL
comparison_df.to_csv('resultados_tcl_modelos.csv', index=False)
print("✓ Resultados TCL salvos em 'resultados_tcl_modelos.csv'")

# Salvar resultados de correlação
corr_results = pd.DataFrame({
    'correlacao': ['Pearson', 'Spearman'],
    'coeficiente': [corr_pearson, corr_spearman],
    'p_valor': [p_pearson, p_spearman],
    'n_observacoes': [len(df), len(df)]
})
corr_results.to_csv('resultados_correlacao.csv', index=False)
print("✓ Resultados de correlação salvos em 'resultados_correlacao.csv'")

print(f"\n✓ Análise completa concluída! ✓")
```
